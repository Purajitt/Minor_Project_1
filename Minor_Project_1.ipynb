```python
#Now read the file
#Since the given file is not in csv format ,it is in txt format so we need to use the following code to tun

with open('/content/hostel_bois.txt', 'r', encoding='utf-8') as file:
    data = file.read()

print(data[:1000])
# Prints the first 1000 characters
```

```python
#Create lists
#Read the Given File
Date=[]
Time=[]
Name=[]
Msg=[]
with open("hostel_bois.txt","r",encoding="utf-8") as file:
    data=file.readlines()
for row in data:
    row=row.strip()
    if not row or "-" not in row:
        continue
    parts=row.split("-", 1)
    timestamp_part=parts[0]
    remaining_text=parts[1]
    time_parts=timestamp_part.split(" ")
    current_date=time_parts[0].replace(",","").strip()
    current_time=time_parts[1].strip()
    if ":" in remaining_text:
        sender,message=remaining_text.split(":", 1)
        Date.append(current_date)
        Time.append(current_time)
        Name.append(sender.strip())
        Msg.append(message.strip())
    else:
        Date.append(current_date)
        Time.append(current_time)
        Name.append("System Notification")
        Msg.append(remaining_text.strip())

#Now we have to convert lists to dataframe
import pandas as pd
df=pd.DataFrame({'Date': Date, 'Time': Time, 'Name': Name, 'Msg': Msg})
df
```

```python
#Now to view data (basic info)
df.info()
```

```python
#Remove all the system messages
#Selects all rows where the sender's name is "System Notification"
system_rows=df[df['Name']=="System Notification"]
#Count the total number of system messages
system_counter=len(system_rows)
# Keep only the rows where the sender is NOT "System Notification" others will
# be removed
df=df[df['Name']!="System Notification"]
print(f"Total number of system messages are: {system_counter}")
```

```python
# Analysing the media messages
#Dictionary to store the number of media messages sent by each person
media_counts={}
#Variable to store the total number of media messages in the chat
total_media_count=0
for index, row in df.iterrows(): #Loop through each dataframe(df)
    msg_text=str(row['Msg'])
    sender_name=row['Name']
    if "<" in msg_text:
        total_media_count=total_media_count + 1 #Increase the total media message count by 1
        if sender_name in media_counts:#Check if the sender already exists in the dictionary
            media_counts[sender_name]=media_counts[sender_name] + 1 # If yes, increase their media count by 1
        else:
           # If the sender is encountered for the first time,
            # initialize their media count to 1
            media_counts[sender_name]=1
print(f"Total number of media entries are: {total_media_count}")
print("Media entries count per person:")
for person, count in media_counts.items():
    print(f"➔ {person}: {count}")
```

```python
# Dictionary to store deleted message count for each person
deleted_counts={}

# Variable to store total deleted messages
total_deleted_count=0

# Loop through each row in the DataFrame
for index,row in df.iterrows():

    # Get the message text and sender name
    msg_text=str(row['Msg']).lower().strip()
    sender_name=row['Name']

    # Check if the message was deleted
    if "this message was deleted" in msg_text:

        # Increase total deleted message count
        total_deleted_count+=1

        # Increase sender's deleted message count
        if sender_name in deleted_counts:
            deleted_counts[sender_name]+=1
        else:
            deleted_counts[sender_name]=1

# Print total deleted messages
print(f"Total number of deleted messages are: {total_deleted_count}")

# Print deleted message count for each person
print("Deleted messages count per person:")

for person,count in deleted_counts.items():
    print(f"➔ {person}: {count}")
```
